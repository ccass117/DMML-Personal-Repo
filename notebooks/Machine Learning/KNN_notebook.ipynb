```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

# DATA
data mining

```python
df = pd.read_excel("Database Pokedex.xlsx")

# Affiche les premières lignes pour vérifier le chargement
# print(df.head())

## Variables
name = df["Pokemon\nName"].tolist()
hp=df["HP"].tolist()
attack=df["Attack"].tolist()
defense=df["Defense"].tolist()
spattack=df["Special\nAttack"].tolist()
spdefense=df["Special\nDefense"].tolist()
speed=df["Speed"].tolist()
total=df["Total"].tolist()
weight=df["Weight (kg)"].tolist()
height=df["Height (m)"].tolist()
type=df["Type I"].tolist()

lentot=len(name)-2
name=name[0:lentot]
hp=hp[0:lentot]
attack=attack[0:lentot]
defense=defense[0:lentot]
spattack=spattack[0:lentot]
spdefense=spdefense[0:lentot]
speed=speed[0:lentot]
total=total[0:lentot]
weight=weight[0:lentot]
height=height[0:lentot]
type=type[0:lentot]



```

allocate of a number for each type

```python
differentypes=["Bug","Dark","Dragon","Electric","Fairy","Fighting","Fire","Flying","Ghost","Grass","Ground","Ice","Normal","Poison","Psychic","Rock","Steel","Water"]
nb = list(range(18))
typesnb=[]

for ty in type:
    for n in range(len(differentypes)):
        if ty==differentypes[n]:
            typesnb.append(nb[n])

print(typesnb)

print(differentypes)
print(nb)
```

Separation of test data and trained data

```python
lengthtrain=int(0.8*len(name))

nametrain=name[0:lengthtrain]
hptrain=hp[0:lengthtrain]
attacktrain=attack[0:lengthtrain]
defensetrain=defense[0:lengthtrain]
spattacktrain=spattack[0:lengthtrain]
spdefensetrain=spdefense[0:lengthtrain]
speedtrain=speed[0:lengthtrain]
totaltrain=total[0:lengthtrain]
weighttrain=weight[0:lengthtrain]
heighttrain=height[0:lengthtrain]
typenbtrain=typesnb[0:lengthtrain]


datatrain=[list(x) for x in zip(hptrain, attacktrain,defensetrain,spattacktrain,spdefensetrain,speedtrain,totaltrain,weighttrain,heighttrain)]
 #without type but type known


nametest=name[lengthtrain:len(name)]
hptest=hp[lengthtrain:len(name)]
attacktest=attack[lengthtrain:len(name)]
defensetest=defense[lengthtrain:len(name)]
spattacktest=spattack[lengthtrain:len(name)]
spdefensetest=spdefense[lengthtrain:len(name)]
speedtest=speed[lengthtrain:len(name)]
totaltest=total[lengthtrain:len(name)]
weighttest=weight[lengthtrain:len(name)]
heighttest=height[lengthtrain:len(name)]
typenbtest=typesnb[lengthtrain:len(name)]




datatest=[list(x) for x in zip(hptest, attacktest,defensetest,spattacktest,spdefensetest,speedtest,totaltest,weighttest,heighttest)]
print(datatest) #without type
```

K Nearest Neighbour function

```python

def euclidean_distance(listetrain,listetest):
    distancecarrée=0
    for i in range(len(listetest)):
        distancecarrée+=(listetest[i]-listetrain[i])**2
    distance=np.sqrt(distancecarrée)
    return distance

def knn_predict_multiclass(X_train, y_train, X_test, k):
    predictions = []
    for test_point in X_test:
        # Compute distances between train point and test point
        distances = [(euclidean_distance(train_point, test_point), label) for train_point, label in zip(X_train, y_train)]
        
        # Sort distances and get the k nearest neighbors
        distances.sort(key=lambda x: x[0])
        k_nearest_neighbors = distances[:k]
        
        # Count occurrences of each class label among the k nearest neighbors
        labels = [label for _, label in k_nearest_neighbors]
        vote_counts = Counter(labels)
        
        # Get class label with the most votes
        most_common_label = vote_counts.most_common(1)[0][0]
        predictions.append(most_common_label)
    
    return predictions
```

```python

# Predict the class for each test point
k = 28
predictions = knn_predict_multiclass(datatrain, typenbtrain, datatest, k)
print("Predicted labels:", predictions)
print(typenbtest)

```

```python
# Création de l'axe des x
x = range(len(typenbtest))

# Création du graphique
plt.figure(figsize=(12,9))
plt.scatter(x, typenbtest, color='blue', marker='o', label='Types')
plt.scatter(x, predictions, color='red', marker='o', label='Predicted types')

plt.title("Predicted types compared to real types with a KNN model")
plt.xlabel("Différent Pokémon")
plt.ylabel("Number allocated to the type")
plt.legend()

# Affichage du graphique
plt.show()
```

```python
erreur=0
for a,b in zip(predictions,typenbtest):
    if a==b:
        erreur+=1
    else:
        erreur+=0
score=erreur/len(predictions)
print(score*100) # the more it is close to 100 the better it is 
```

```python
score_rate = []

# Will take some time
for i in range(1,40):
  erreur=0
  pred_i = knn_predict_multiclass(datatrain, typenbtrain, datatest, i)
  for a,b in zip(pred_i,typenbtest):
    if a==b:
      erreur+=1
    else:
      erreur+=0
  score=100*erreur/len(predictions)
  score_rate.append(score)

```
